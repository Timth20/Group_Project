```python
#Import librarie

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import math
sns.set_style('whitegrid')
```

```python
#Import data: the indicators with continuous multi-year changes

ruanke = pd.read_csv('ruanke.csv')
```

```python
QS = pd.read_csv('QS.csv')
```

```python
THE = pd.read_csv('THE.csv')
```

```python
#The change of house price in HK from 2020 to 2022

plt.figure(figsize=(16,8))
sns.lineplot(x='Rank', y='Course', data=ruanke)
plt.xticks(rotation=-20)
#plt.xlabel('year',fontsize=16,color='royalblue')
#plt.ylabel('house price',fontsize=16,color='royalblue')
#plt.title('The change of house price in HK from 2020 to 2022',size=20,color='royalblue',fontweight='bold')
plt.show()
```

```python
graph = ruanke.groupby('Region')['Course'].mean()

# 

graph_sorted = graph.sort_values(ascending=False)
graph_sorted.plot(kind='bar')
plt.xlabel('Region')
plt.ylabel('Average Course')
plt.title('Average Course by Region')
plt.show()
```
